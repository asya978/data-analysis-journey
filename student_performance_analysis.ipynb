# Student Performance Data Analysis

This project explores a simple dataset to practice:

- Data creation
- Basic statistics
- Data visualization
- Interpretation of results

Tools used:
- Python
- Pandas
- Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample dataset
data = {
    "Study Hours": [7, 5, 2, 6, 5, 6, 5,],
    "Topic Studied": [5, 3, 3, 4, 4, 6, 4]
}

df = pd.DataFrame(data)

df
```

```python
print("Average Study Hours:", df["Study Hours"].mean())
print("Average Topic Studied:", df["Topic Studied"].mean())
```

```python
plt.scatter(df["Study Hours"], df["Topic Studied"])
plt.title("Study Hours vs Topic Studied")
plt.xlabel("Study Hours")
plt.ylabel("Topic Studied")
plt.show()
```

## Interpretation

There is a positive relationship between study hours and topic studied.
As study time increases,  topics studied improves.

This suggests a correlation between effort and diversity.
